```python
url='https://dxy.com/app/i/content/detail?id=24677&content_status=1&display_course_entry=1'
```

```python
QA_dict={}
try:
    response = requests.get(url,headers=headers)#请求网页内容
    if response.status_code == 200:#判断是否正确响应
        for i in range(7):
            page=response.json().get('data').get('items')[0]['article'][i]
            for j in range(len(page)):
                Q=re.sub("<(.*?)>",'',page['detail_search'][j]['title'].replace('\n',''))[2:]
                A=re.sub("<(.*?)>",'',page['detail_search'][j]['content'].replace('\n',''))
                QA_dict[Q]=A
except requests.ConnectionError as e:
    print('Error', e.args)
```
